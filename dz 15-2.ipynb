Написать скрипт, который будет проверять список e-mail адресов на утечку при помощи сервиса Avast Hack Ckeck https://www.avast.com/hackcheck/. Список email-ов задаем переменной в начале кода:

EMAIL = [xxx@x.ru, yyy@y.com]

В итоге должен формироваться датафрейм со столбцами: <почта> - <дата утечки> - <источник утечки> - <описание утечки>

Подсказка: сервис работает при помощи "скрытого" API. Внимательно изучите post-запросы.


```python
import requests
```

```python
import pandas as pd
```

```python
import json
```

```python
def breach_email(mail):
    #email = ['xxx@xx.com']
    url = 'https://identityprotection.avast.com/v1/web/query/site-breaches/unauthorized-data?'
    headers = {
        'Vaar-Version' : '0',
        'Vaar-Header-App-Product': 'hackcheck-web-avast'
    }
    params = {
        'emailAddresses': mail
    }
    r = requests.post(url, data=json.dumps(params), headers=headers)
    r.json()
    df = pd.DataFrame(r.json()['breaches'])
    df = df.transpose().reset_index()
    df = df.rename(columns={'index': 'breach'})
    return df.head()
#breach_email(['aaa@gmail.com'])
```

```python
email = ['xxx@xx.com']
breach_email(email)
```
